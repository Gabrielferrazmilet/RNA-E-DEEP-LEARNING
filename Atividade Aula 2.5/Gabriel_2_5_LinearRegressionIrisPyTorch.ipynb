# Regressão Linear com PyTorch

## Objetivo

Este exemplo utiliza regressão linear para estimar o comprimento das sépalas da íris a partir do seu comprimento das pétalas.
Utiliza-se
- o cálculo do gradiente automático do PyTorch, porém
- a função de perda e
- atualização dos parâmetros é feita de forma explícita.
- A rede é criada com uma camada nn.Linear()

![alt text](https://raw.githubusercontent.com/vcasadei/images/master/RegressaoLinear_Otimizacao.png)

## Importação dos pacotes

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
from sklearn.datasets import load_iris

import torch
from torch import nn, optim
from torch.autograd import Variable

torch.manual_seed(1234)
```

## Leitura dos dados

```python
iris = load_iris()
data = iris.data[iris.target==1,::2]  # comprimento das sépalas e pétalas, indices 0 e 2

x_train = data[:,0:1].astype(np.float32)
y_train = data[:,1:2].astype(np.float32)

n_samples = x_train.shape[0]
print('x_train.shape:',x_train.shape, x_train.dtype)
print('y_train.shape:',y_train.shape, y_train.dtype)

print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
```

### Normalização dos dados

```python
x_train -= x_train.min()
x_train /= x_train.max()
y_train -= y_train.min()
y_train /= y_train.max()
```

```python
print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
```

```python
x_train_bias = np.hstack([np.ones(shape=(n_samples,1)), x_train])
```

```python
x_train_bias
```

```python
print('x_train_bias.shape:',x_train_bias.shape, x_train_bias.dtype)
```

```python
x_train_bias = torch.FloatTensor(x_train_bias)
y_train      = torch.FloatTensor(y_train)
print(y_train)
```

## Model

![alt text](https://raw.githubusercontent.com/vcasadei/images/master/RegressaoLinearNeuronio.png)

```python
#Inicialização usando toch.nn.Linear

model = torch.nn.Linear(in_features=2, out_features=1, bias=False)
```

### Verificando a inicialização dos parâmetros

```python
model.weight.data = torch.zeros(1,2)
torch.nn.init.uniform(model.weight.data, -0.1, 0.1)
model.weight
```

### Testando o predict da rede

```python
xx = torch.ones((5,2))
xx
```

```python
model(Variable(xx))
```

## Treinamento

```python
num_epochs = 100
learning_rate = 0.5
for epoch in range(num_epochs):
    #inputs = Variable(x_train_bias)
    #target = Variable(y_train)
    inputs = x_train_bias
    target = y_train

    # forward - predict
    y_pred = model(inputs)

    #loss cálculo da função de perda
    loss = torch.mean((y_pred - target)**2)

    # backward propagation - cálculo gradient
    loss.backward()

    # gradiente descendente
    model.weight.data = model.weight.data - learning_rate * model.weight.grad.data

    # zeragem parâmetros próximo cálculo do gradiente
    model.weight.grad.data.zero_()

    # verbose
    if (epoch+1) % 20 == 0:
        print('Epoch[{}/{}], loss: {:.6f}'
              .format(epoch+1, num_epochs, loss.data))
```

## Avaliação

```python
y_pred = model(Variable(x_train_bias))
plt.plot(x_train, y_train.numpy(), 'ro', label='Original data')
plt.plot(x_train, y_pred.data.numpy(), 'kx-', label='Fitting Line')
plt.show()
```

### Parâmetros ótimos, solução analítica

```python
x_bias = x_train_bias
y = y_train
x_bias_t = torch.t(x_bias)
w_opt = (torch.inverse(x_bias_t.mm(x_bias)).mm(x_bias_t)).mm(y)
print(torch.t(w_opt))
```

# Exercícios

- 1) Por que no final da varredura do laço de treinamento, a perda é impressa utilizando
   `loss.data`? Por que precisa do `data`?
   

    R:

Como a variável loss carrega o histórico das operações matemáticas para realizar a retropropagação, precisamos do .data para filtrar esse histórico e apresentar no log apenas o número real que representa o erro final do treinamento\
    obs: em versões mais recentes do pytorch também é comum usar o ".item()" para essa mesma finalidade de extrair o valor escalar

```python
model.eval()
with torch.no_grad():
    y_pred_train = model(x_train_bias)   # ou x_test_bias para avaliar no teste
    mse_final = torch.mean((y_pred_train - y_train)**2).item()
print(f"MSE final (treino, usando pesos treinados): {mse_final:.6f}")
```

- 3) Crie duas listas `w0_list` e `w1_list` para guardar a evolução dos valores dos
   parâmetros durante o laço de treinamento.
   Modifique o exemplo do gráfico tipo **scatter plot** para visualizar a variação
   dos parâmetros do modelo duas dimensões.
   Marque neste gráfico o valor ótimo do parâmetro dado pela solução
   analítica.

```python

import torch
import matplotlib.pyplot as plt

# listas para armazenar evolução dos pesos
w0_list = []
w1_list = []

num_epochs = 100
learning_rate = 0.5


for epoch in range(num_epochs):
    inputs = x_train_bias            # (N,2) -> [1, x]
    target = y_train                 # (N,1) ou (N,)

    y_pred = model(inputs)

    loss = torch.mean((y_pred - target)**2)

    loss.backward()

    with torch.no_grad():
        model.weight.data = model.weight.data - learning_rate * model.weight.grad.data
        # gravar os pesos atuais (w0 = coef sobre a coluna de 1s, w1 = coef sobre x)
        w0_list.append(float(model.weight.data[0,0].item()))
        w1_list.append(float(model.weight.data[0,1].item()))

        # zerar gradientes
        model.weight.grad.data.zero_()

    # print opcional
    if (epoch+1) % 20 == 0:
        print(f"Epoch {epoch+1}/{num_epochs} - loss: {loss.item():.6f}")


# x_train_bias tem shape (N,2): [1, x]
X_design = x_train_bias
y_vec = y_train.view(-1,1)

# usar pinverse
theta_opt = torch.pinverse(X_design).mm(y_vec)

bias_opt = float(theta_opt[0].item())
w1_opt   = float(theta_opt[1].item())

print("Solução analítica (theta):", torch.t(theta_opt))
print(f"bias_opt = {bias_opt:.6f}, w1_opt = {w1_opt:.6f}")

# --- Plot da trajetória dos pesos e do ponto ótimo ---
plt.figure(figsize=(6,6))
if len(w0_list) > 0:
    plt.scatter(w0_list, w1_list, s=30, label='iterações')
    plt.plot(w0_list, w1_list, alpha=0.6)
else:
    print("Aviso: w0_list/w1_list vazias (verifique se o laço de treino foi executado).")

# ponto ótimo (extraído de theta_opt)
plt.scatter([bias_opt], [w1_opt], marker='*', s=300, label='ótimo analítico')

plt.xlabel("w0 (coef sobre 1s / intercept)")
plt.ylabel("w1 (coef sobre x / slope)")
plt.title("Trajetória dos parâmetros (w0, w1)\nEstrela = solução analítica")
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
